```python
import pandas as pd
import json
```

```python
symbol = "F"

with open(f"../../data/balance_sheets/{symbol}_balance_sheet.json", "r") as f:
    bs_data = json.load(f)

with open(f"../../data/income_statements/{symbol}_income_statement.json", "r") as f:
    income_statement_data = json.load(f)

with open(f"../../data/cashflow_statements/{symbol}_cashflow_statement.json", "r") as f:
    cashflow_statement_data = json.load(f)

df_bs = pd.json_normalize(bs_data["quarterly_reports"])
df_inc = pd.json_normalize(income_statement_data["quarterly_reports"])
df_cf = pd.json_normalize(cashflow_statement_data["quarterly_reports"])

display(df_bs.head())
display(df_inc.head())
display(df_cf.head())
```

```python
print(df_bs.dtypes)
print(df_inc.dtypes)
print(df_cf.dtypes)
```

```python
import altair as alt

# Ensure your dates are datetime type
df_bs["fiscal_date_ending"] = pd.to_datetime(df_bs["fiscal_date_ending"])

line_assets = alt.Chart(df_bs).mark_line(color="blue").encode(
    x=alt.X("fiscal_date_ending:T"),
    y=alt.Y("total_assets:Q")
)

line_liabilities = alt.Chart(df_bs).mark_line(color="red").encode(
    x=alt.X("fiscal_date_ending:T"),
    y=alt.Y("total_liabilities:Q")
)

layered_chart = alt.layer(line_assets, line_liabilities)
layered_chart
```
